## Load Data

Kita akan menggunakan python untuk memasukkan data dan menampilkannya dengan tabel memanfaatkan library **pandas**

**NB : Jangan panik jika Anda menemukan tanda ... pada tabel pandas. Tanda tersebut merupakan indikasi bahwa pandas hanya menampilkan sebagian data ke interface**.

```python
import pandas as pd #import library pandas dan simpan dengan nama pd

data = pd.read_csv('test.csv') #load data 'test.csv'
data # tampilkan data
```

## Pahamai Data

Sekarang, setelah data kita dapatkan, mari coba kita perhatikan data penumpang kapal Titanic tersebut. Data tersebut menjelaskan informasi tentang penumpang (baris) pada setiap kolom. Data pada kolom merupakan data kategorial yang dapat berupa nilai kuantitatif ataupun kualitatif (label). Akan tetapi tidak ada analisis yang dapat kita lakukan jika mengubungkan baris dengan kolom. Informasi hanya dibaca per baris. Model data seperti ini dimanakan dengan _**one-way data / one-way table**_. 

Pada data tersebut, kolom-kolom disebut dengan variable dari data. Pertanyaannya bagaimana jika susunan data diubah? Penumpang menjadi kolom, informasi tentang penumpang menjadi baris. Data tersebut tetap data one-way. Penyusunan data dari vertikal menjadi horizontal, tergantung dengan kebutuhan

> ### Best Practice : Jika jumlah variabel lebih banyak dari datanya, maka lebih baik jika data ditampilkan dalam format horizontal. Variable sebagai baris dan data sebagai kolom

## Merepresentasikan One-way Data / One-way Table

Salah dua cara untuk merepresentasikan one-way data / one-way table adalah dengan menggunakan bar chart atau pie chart. Bar chart dapat digunakan untuk menampilkan informasi terkait dengan frekuensi data. Sedangkan pie chart dapat digunakan untuk menampilkan prosentase dari data.

### Studi Kasus 1 : Berapa Jumlah Penumpang untuk Tiap Kategori Kelas?

Mari selesaikan dengan menggunakan python!

```python
# PASTIKAN BLOCK KODE LOAD DATA SUDAH DIJALANKAN SEBELUM MENJALANKAN BLOK KODE INI
# Jalankan sekali kode ini, jika pada nilai PClass menjadi NaN atau ada error lain,
# jalankan kembali block kode untuk load data

# Langkah 1 : Ubah nilai Pclass menjadi label 
data['Pclass'] = data['Pclass'].map({1:'1st Class', 2:'2nd Class', 3:'3rd Class'}) # mapping
data # cetak data
```

```python
# Langkah 2 : Hitung jumlah tiap kelas

jml_kelas = data['Pclass'].value_counts()
jml_kelas
```

```python
# Langkah 3 : Import Library matplotlib
import matplotlib.pyplot as plt # import matplotlib untuk plotting
```

```python
# Langkah 4 : Lakukan plotting :)

jml_per_kelas = [jml_kelas['1st Class'], jml_kelas['2nd Class'], jml_kelas['3rd Class']] # data jumlah penumpang
label_kelas = ['1st Class', '2nd Class', '3rd Class'] # label kelas

plt.bar(label_kelas, jml_per_kelas) # fungsi plot
plt.title('Jumlah Penumpang Per Kelas Kapal Titanic') # tambahkan judul
plt.xlabel('Nama Kelas') # keterangan sumbu x
plt.ylabel('Jumlah (orang)') # keterangan sumbu y
plt.show() # tampilkan plot
```

#### Bagimana jika kita ingin memunculkan dalam bentuk prosentase ?

Untuk menampilkan data dalam bentuk prosentase, _**pie chart**_ merupakan salah satu pilihan yang tepat. Pie chart dapat memberikan pembaca mengenai bagian data dibandingan dengan keseluruhan data.

```python
# Membuat Pie Chart

jml_per_kelas = [jml_kelas['1st Class'], jml_kelas['2nd Class'], jml_kelas['3rd Class']] # data jumlah penumpang
label_kelas = ['1st Class', '2nd Class', '3rd Class'] # label kelas

fig, ax = plt.subplots()
ax.pie(jml_per_kelas, label_kelas, autopct='%1.1f%%') # masukkan data
ax.axis('equal') # membagi data sesuai dengan ukuran dan membentuk lingkaran
plt.show()
```
